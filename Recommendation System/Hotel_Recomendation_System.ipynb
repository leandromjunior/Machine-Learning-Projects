```python
# Creating a function to convert the strings of list into a normal list/series (removing the punctuation in general)
def remove_punctuation(col):
  col = col[0]
  if(type(col) != list):
    return "".join(literal_eval(col))
  else:
    return col
```

```python
# Applying the function to the Tags column and getting rid of its commas and apostrophes
new_df['Tags'] = new_df[['Tags']].apply(remove_punctuation, axis=1)
```

```python
new_df.head()
```

```python
new_df['Tags'] = new_df['Tags'].str.lower()
new_df['Countries'] = new_df['Countries'].str.lower()
```

```python
def recommender(countrySail, keywords):
  keywords = keywords.lower()
  # Transforming the sentences into words
  word_tokenize(keywords)

  stop_words = stopwords.words('english')

  lemm = WordNetLemmatizer()

  filtered = {word for word in keywords if not word in stop_words}
  filtered_set = set()
  for fs in filtered:
    filtered_set.add(lemm.lemmatize(fs))

  # location
  country = new_df[new_df['Countries'] == countrySail.lower()]
  country = country.set_index(np.arange(country.shape[0]))

  list1 = []
  list2 = []
  cosineList = []

  for i in range(country.shape[0]):
    temp_token = word_tokenize(country['Tags'][i])
    temp_set = [word for word in temp_token if not word in stop_words]
    temp2_set = set()

    for s in temp_set:
      temp2_set.add(lemm.lemmatize(s))

    vector = temp2_set.intersection(filtered_set)
    cosineList.append(len(vector))

  # Applying cosine similarity
  country['similarity'] = cosineList
  country = country.sort_values(by='similarity', ascending=False)
  country.drop_duplicates(subset="Hotel_Name", keep="first", inplace=True)
  country.sort_values('Average_Score', ascending=False, inplace=True)
  country.reset_index(inplace=True)

  return country[['Hotel_Name', 'Average_Score', 'Hotel_Address']].head(10)
```

```python
recommender('NL', 'Business trip Couple 2 nights')
```
